```python
## Step 3
## Archive previous results - Better to run after each execution
from glob import glob
import os
import time

def archive_files():
    timestr = time.strftime("%Y-%m-%d_%H%M%S")
    location = "/content/drive/MyDrive/Projects/mri-variationalnetwork"
    archive = os.path.join(location, "archives", timestr)

    print(f"Archiving files to {archive} ...")
    os.makedirs(archive, exist_ok=True)
    result_files = glob(os.path.join(location, "result-*"))
    for file in result_files:
      os.rename(file, os.path.join(archive, os.path.basename(file)))
    print("Archiving complete.")

archive_files()
```
